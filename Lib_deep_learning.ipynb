# ANN

```python
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from tensorflow.keras.wrappers.scikit_learn import KerasRegressor 
from sklearn.model_selection import GridSearchCV
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Build Keras model 2 hidden layers
def build_model_2layers(units_1,units_2,input_shape,activation_func,activation_func_output,optimizer,loss,metrics):
  model = keras.Sequential()
  # the hidden layers
  model.add(layers.Dense(units=units_1, input_shape=[input_shape], activation=activation_func))
  model.add(layers.Dense(units=units_2, activation=activation_func))
  # the output layers
  model.add(layers.Dense(units=1, activation= activation_func_output))
  model.compile(optimizer=optimizer, loss=loss, metrics=metrics)
  return model, model.summary()
```

```python
# Fit model with Early stopping 
def fit_model_earlyStopping(model,X_train, y_train,X_valid, y_valid, epochs):
  early_stopping = EarlyStopping(
                                  min_delta=0.001,    
                                  patience=10,       
                                  restore_best_weights=True)
  history = model.fit(X_train, y_train, 
                    validation_data=(X_valid, y_valid), # or validation_split=0.2
                    batch_size=32, #default=32
                    epochs=200,
                    callbacks=[early_stopping], 
                    verbose=1) #turn on training log
  history_df = pd.DataFrame(history.history)
  print(history_df.tail())
  return model
```
